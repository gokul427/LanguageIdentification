```python
import import_ipynb
import os
import re
import operator
import nltk
from nltk.tokenize import word_tokenize as wt
from nltk.util import ngrams,trigrams

from tkinter import *
from tkinter import ttk

from miniEng import ClassEng
from miniGerman import ClassGer
from miniFrench import ClassFrench
from miniPortuguese import ClassPor
from miniSpan import ClassSpan


def calculate(*args):
  
    value=inp.get()
      
    f=open(value,"r",encoding='utf-8')
        
        #reading contents
    if f.mode=="r":
        data=f.read()
        
    r=func(data)
    
    output.set(r) 
    
root = Tk()
root.title("Language Identification")
mainframe = ttk.Frame(root)

mainframe.grid(column=0, row=0, sticky=(N, W, E, S))

inp= StringVar()
output = StringVar()

inp_entry = ttk.Entry(mainframe, width=30, textvariable=inp)
inp_entry.grid(column=3, row=2, sticky=(W, E))

ttk.Label(mainframe, textvariable=output).grid(column=2, row=3, sticky=(W, E))
ttk.Button(mainframe, text="Compute", command=calculate).grid(column=3, row=4, sticky=W)

photo = PhotoImage(file="langs.gif")
ttk.Label(mainframe,image=photo).grid(column=4, row=1)
ttk.Label(mainframe, text="Enter the file name").grid(column=1, row=2,sticky="W")
ttk.Label(mainframe, text="Language is").grid(column=1, row=3,sticky="W")


inp_entry.focus()

root.bind('<Return>', calculate)

root.mainloop()

```

```python
def func(data):
  #cleaning the data by:
   #converting upper case letters to lower case letters
   #replacing the non-word characters(punctuations) with space
    data1=data.lower()
    clean_text = re.sub(r"[^\w]+", " ", data1)
        
    #print(clean_text+"\n")
        
    #Creating tokens using word tokenizer of nltk
    dat=wt(clean_text)
        #print(dat)
        #print("\n")
        
    #Trigrams for each word
    grams_list=[]
    for t in dat:
        token_trigrams = (trigrams((t)))
        for x in token_trigrams:
            grams_list.append(x)
        #print(grams_list[:15])
        #print("\n")
        
        
    #Concatinating 
    concat_list=[]
    
    for i in range(len(grams_list)):
        concat_list.append("".join(grams_list[i]))
        
        
        #print(concat_list[:15])
        
    #Counting the frequency of each trigram and storing them to dictionary
        
    d = {}

    for item in concat_list:
        if item in d:
            d[item] = d.get(item)+1
        else:
            d[item] = 1
        '''for k, v in  d.items():
            print(k, v)
        print("\n")'''
        
    #Sorting trigrams based on thier frequency in descending order
     
    ngrams_statistics_sorted = sorted(d.items(),key=operator.itemgetter(1),reverse=True)
    #print(ngrams_statistics_sorted[:15])
    #print("\n")
    #extracting only trigrams and creating a list
    temp=[i[0] for i in ngrams_statistics_sorted]
        #print(temp[:15])
    r=result(temp)
    return r   
        
#function to compare two lists,find out of measure and return their sum    
def comp(list1,list2):
    flag=0
    s=[]
    for i in range(len(list2)):
        flag=0
        for j in range(len(list1)):
            if list2[i]==list1[j]:
                s.append(abs(i-j))
                flag+=1
        if(flag==0):
            s.append(999)
    total=0
    for i in s:
        total+=i
    return total

def result(testList):
    engList=ClassEng.func()
    gerList=ClassGer.func()
    frenchList=ClassFrench.func()
    porList=ClassPor.func()
    spanList=ClassSpan.func()
    lang=["ENGLISH","GERMAN","FRENCH","PORTUGUESE","SPANISH"]
    rank=[]
    lists=[engList,gerList,frenchList,porList,spanList]
    for names in lists:
        rank.append(comp(names,testList))
    file=open("opRank.txt",'w+')
    file.write("\n\n'ENGLISH','GERMAN','FRENCH','PORTUGUESE','SPANISH'\n ")
    i=0
    for item in rank :
        file.write("{0}, ".format(item))
    file.close()  
        #find index of minimum rank
    ind=rank.index(min(rank))
    file=open("opRank.txt",'a+')
    file.write("\n\n minimum ranked language is %s"%lang[ind])
    file.close()
    return lang[ind]
            
        
    
```
